```python
remove=["symboling","carheight","stroke","compressionratio","peakrpm","doornumber","enginelocation"]
df=df.drop(remove,axis=1)
```

```python
df.head()
```

```python
df=df.drop("brand",axis=1)
```

```python
df.head()
```

```python
sig_col = ['price','enginetype','fueltype', 'aspiration','carbody','cylindernumber', 'drivewheel',
            'wheelbase','curbweight', 'enginesize','fuelsystem' ,'boreratio','horsepower', 
                    'citympg','highwaympg', 'carlength','carwidth']

n_df=pd.get_dummies(df, drop_first=True)
```

```python
pd.set_option("display.max_columns", None)
n_df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data=["wheelbase","carlength","carwidth","curbweight","enginesize","boreratio","horsepower","citympg",
             "highwaympg", "price"]
n_df[scaled_data]= scaler.fit_transform(n_df[scaled_data])
```

```python
n_df
```

```python
X=n_df.drop("price",axis=1)
y=n_df["price"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 42) 
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(fit_intercept = True)
regressor.fit(x_train, y_train)
```

```python
print('Linear Model Coefficients (m)', regressor.coef_)
print('Linear Model Coefficients (b)', regressor.intercept_)
```

```python
y_predict = regressor.predict(x_test)
y_predict 
```

```python
a=pd.DataFrame(regressor.coef_.tolist())
b=pd.DataFrame(X.columns.to_list())
c=pd.concat([a,b],axis=1)
c.columns=['values','features']
c
```

```python
k = x_test.shape[1]
n = len(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt

RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_predict)) , '.3f'))
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2) * (n-1)/(n-k-1)
MAPE = np.mean( np.abs( (y_test-y_predict) / y_test  )  )*100 
```

```python
print('RMSE = ', RMSE, '\nMSE =', MSE , '\nMAE =', MAE, '\nR2 =', r2, '\nAdjusted R2 =', adj_r2, '\nMAPE =', MAPE)
```

```python
plt.scatter(x=y_test, y=y_predict,c='crimson')
p1 = max(max(y_predict), max(y_test))
p2 = min(min(y_predict), min(y_test))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('Actual')
plt.ylabel('Prediction')
plt.title('Actual vs. Multi-Linear reg Prediction')
```
